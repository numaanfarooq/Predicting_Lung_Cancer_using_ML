```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(4, 4, i+1)
    sns.histplot(df[column], bins=20, kde=True)
    plt.title(column)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(26, 10))
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(4, 4, i+1)
    if column == 'AGE':
        sns.histplot(x=column, hue='LUNG_CANCER', data=df, palette='Set2', bins=20, kde=True)
        plt.xlim(0, 100)  # Adjust the x-axis limits to focus on a specific age range
    else:
        sns.countplot(x=column, hue='LUNG_CANCER', data=df, palette='Set2')
    plt.title(f'Distribution of {column} by LUNG_CANCER', fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(title='LUNG_CANCER', fontsize=12)
plt.tight_layout()
plt.show()
```
