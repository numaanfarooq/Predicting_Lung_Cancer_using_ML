```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import seaborn as sns
```

```python
df = pd.read_csv('data/lungcancer.csv')
```

##Dataset Information

Gender: Gender of the patient (M: Male, F: Female)

Age: Age of the patient

Smoking: Smoking status (YES=2, NO=1)

Yellow Fingers: Presence of yellow fingers (YES=2, NO=1)

Anxiety: Anxiety levels (YES=2, NO=1)

Peer Pressure: Influence of peer pressure (YES=2, NO=1)

Chronic Disease: Presence of chronic diseases (YES=2, NO=1)

Fatigue: Experience of fatigue (YES=2, NO=1)

Allergy: Presence of allergies (YES=2, NO=1)

Wheezing: Wheezing symptoms (YES=2, NO=1)

Alcohol Consuming: Alcohol consumption status (YES=2, NO=1)

Coughing: Presence of coughing (YES=2, NO=1)

Shortness of Breath: Experience of shortness of breath (YES=2, NO=1)

Swallowing Difficulty: Difficulty swallowing (YES=2, NO=1)

Chest Pain: Presence of chest pain (YES=2, NO=1)

Lung Cancer: Lung cancer diagnosis (YES, NO)

```python
#Visualize the data

#Visualize the gender Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='GENDER', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], bins=10, kde=False)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='SMOKING', data=df)
plt.title('Smoking_Status Distribution')
plt.show()
```

#Insights so far


Gender Distribution:
The dataset is fairly balanced between male (162) and female (147) patients. The dataset shows a slightly higher number of male patients (52%) compared to female patients (48%)

Age Distribution:
The average age is approximately 62.67 years, with a standard deviation of 8.21 years.
The age range is from 21 to 87 years.


A significant portion of the patients are smokers.
Other health attributes (yellow fingers, anxiety, peer pressure, chronic disease, etc.) are recorded as binary values, indicating the presence (2) or absence (1) of these conditions.
